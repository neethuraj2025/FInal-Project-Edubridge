## 7: Model Creation

Splitting the dataset
 <ul>

```python
x = d.drop(['Type'],axis=1)
y = d['Type']
```

###  Splitting dataset to train and test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=27)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Training the data with ML approaches
<ul>
  <li>Logistic Regression</li>
  <li>KNN</li>
  <li>Random Forest Classifier</li>
  <li>Naive Bayes Classifier</li>
  <li>AdaBoost Classifier</li>
  <li>Decisiontree Classifier</li>
</ul>



ML approach Functions
<ul>
  <li>Model fitting</li>
  <li>Predicting using test</li>
  <li>Accuracy score</li>
  <li>confusion matrix</li>
  <li>time taken by each model</li>
</ul>

```python
def model_create(model, x_train, y_train, x_test, y_test):
    intial_time=time.time()
    model.fit(x_train,y_train.ravel())
        
    y_pred = model.predict(x_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy = {}".format(accuracy))
    
    time_taken = time.time()-intial_time
    print("Time taken = {}".format(time_taken))
    
    print(classification_report(y_test,y_pred,digits=5))
    
    plot_confusion_matrix(model, x_test, y_test,cmap=plt.cm.pink, normalize = 'all')                   
    
    return model, accuracy, time_taken
```

```python
model_LR=LogisticRegression()
model_LR, accuracy_LR,  time_LR= model_create(model_LR, x_train, y_train, x_test, y_test)
```

```python
model_DC=DecisionTreeClassifier()
model_DC, accuracy_DC, time_DC = model_create(model_DC, x_train, y_train, x_test, y_test)
```

```python
model_RF=RandomForestClassifier()
model_RF, accuracy_RF, time_RF = model_create(model_RF, x_train, y_train, x_test, y_test)
```

```python
model_NB=GaussianNB()
model_NB, accuracy_NB, time_NB = model_create(model_NB, x_train, y_train, x_test, y_test)
```

```python
model_KNN=KNeighborsClassifier(n_neighbors=5)
model_KNN, accuracy_KNN, time_KNN = model_create(model_KNN, x_train, y_train, x_test, y_test)
```

```python
model_AB=AdaBoostClassifier(n_estimators=50,learning_rate=1)
model_AB, accuracy_AB, time_AB = model_create(model_AB, x_train, y_train, x_test, y_test)
```
